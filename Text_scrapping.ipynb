```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://insights.blackcoffer.com/what-jobs-will-robots-take-from-humans-in-the-future/"
```

```python
Page = requests.get(url)
```

```python
Page
```

```python
Data = BeautifulSoup(Page.content, 'html.parser')
```

```python
Data
```

```python
title = Data.find('h1').text
```

```python
title
```

```python
BoldHeading = Data.findAll('h3')
```

```python
BoldHeading
```

```python
SubHeading = Data.findAll('strong')
SubHeading
```

```python
Paragraph = Data.findAll('p')
Paragraph
```

```python
Bullets = Data.findAll('li')
Bullets
```

```python
Data_2 = """ """
```

```python
Data_2 += title
Data_2 += '\n'
```

```python
i =4
while(i<5):
    Data_2 += SubHeading[i].text
    Data_2 += '\n'
    i+=1
```
